# Imports

```
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
from sklearn.metrics import confusion_matrix

import math
```

# Downloading Data

```
from tensorflow.examples.tutorials.mnist import input_data
data = input_data.read_data_sets('data/MNIST/', one_hot = True)
```

# Exploring Data

```
print("Size of:")
print("- Training-set:\t\t{}".format(len(data.train.labels)))
print("- Validation-set:\t{}".format(len(data.validation.labels)))
print("- Test-set:\t\t{}".format(len(data.test.labels)))
```

# **Data Dimensions variables**

```
img_size = 28 #images are 28 Pixels in each direction
```

```
img_size_flat = img_size * img_size #images are stored in one dimensional array of this length
```

```
img_shape = (img_size, img_size)
```

```
num_classes = 10
```

# Placeholders and Varaibles

```
X = tf.placeholder(tf.float32, [None, img_size_flat]) 
#placeholder in tensorflow hold the input data
#placeholder variable for input ....tf.float32 is the data type....
#[None, img_size_flat] means it is a 2-d Tensor....None means there can be arbitrary number of rows....no of columns  = img_size_flat for each row
```

```
y = tf.placeholder(tf.float32, [None, num_classes]) #one hot encoded classes i.e, 10 columns for each class
```

```
y_cls = tf.placeholder(tf.int64, [None])      #normal interger classes
```

# Deep NN Model

```
L = 200
M = 100
N = 60
O = 30
```

```
W1 = tf.Variable(tf.truncated_normal([img_size_flat, L], stddev = 0.1))  #truncated_normal means small random values
b1 = tf.Variable(tf.zeros([L]))

W2 = tf.Variable(tf.truncated_normal([L, M], stddev = 0.1))
b2 = tf.Variable(tf.zeros([M]))

W3 = tf.Variable(tf.truncated_normal([M, N], stddev = 0.1))
b3 = tf.Variable(tf.zeros([N]))

W4 = tf.Variable(tf.truncated_normal([N, O], stddev = 0.1))
b4 = tf.Variable(tf.zeros([O]))

W5 = tf.Variable(tf.truncated_normal([O, num_classes], stddev = 0.1))
b5 = tf.Variable(tf.zeros([num_classes]))
```

```
Y1 = tf.nn.sigmoid(tf.matmul(X, W1)+b1)
Y2 = tf.nn.sigmoid(tf.matmul(Y1, W2)+b2)
Y3 = tf.nn.sigmoid(tf.matmul(Y2, W3)+b3)
Y4 = tf.nn.sigmoid(tf.matmul(Y3, W4)+b4)
Ylogits = tf.matmul(Y4, W5) + b5
y_pred = tf.nn.softmax(Ylogits) 
```

# Loss

```
cross_entropy = -tf.reduce_sum(y*tf.log(y_pred))
```

# Accuracy

```
is_correct = tf.equal(tf.argmax(y_pred, 1), tf.argmax(y, 1))
```

```
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
```

# Gradient

```
optimizer = tf.train.GradientDescentOptimizer(0.003) #0.003 is learning rate
```

```
train_step = optimizer.minimize(cross_entropy)
```

# Starting Tensorflow Session

```
sess = tf.Session()
```

```
init = tf.global_variables_initializer()
sess.run(init)
```

# Training Loop

```
for i in range(24700):
  
  #loading batch of images and labels
  batch_X, batch_y = data.train.next_batch(100)
  train_data = {X: batch_X, y : batch_y}
  
  #training
  sess.run(train_step, feed_dict = train_data)  #train step is the step that computed gradients
  
  #success on training data
  train_accuracy, train_cross_entropy = sess.run([accuracy, cross_entropy], feed_dict = train_data)
  
  #success on test data
test_data = {X: data.test.images, y: data.test.labels}
test_accuracy, test_cross_entropy = sess.run([accuracy, cross_entropy], feed_dict = test_data)
  
```

```
def print_accuracy():
  acc = sess.run(accuracy, feed_dict = test_data)
  
  print('Accuracy on test-set : {0: .1%}'.format(acc))
```
